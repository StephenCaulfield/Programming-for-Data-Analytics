## 1.2 Reasons for Simulating Data

<p>I will detail the advantages and disadvantages of simulating a data set to imitate social phenomon.</p>

#### Advantages

* It gives a more detailed and unique perspective of the phenomenon, allowing us to precisely analyse what the outcomes are instead of making baseless presumptions.
    
* It can predict how people will behave in potentially similar events.
    
* It can help professionals develop law's, new technologies and philosophies that help improve our lives(e.g. increase survival rate of ship passengers)
    
* Simulating these events is risk-free(No need to sink a ship for data analysis)

#### Disadvantages

* Researcher-induced bias can potentially cause wildly different interpretations of the data.
    
* A subjective analysis is necessary to derive meaning from the data set in social phenomenons.
    
* Social phenomonons can change over time due to cultural and technological differences which can make these simulations appear dated very quickly.

## 2.2 Explaining the dataset
<span style="color:blue">Blue</span> = Relevant to analysis. |<span style="color:red">Red</span> = Not Relevant to analysis.|<span style="color:orange">Orange</span> = May be somewhat releavant to analysis.

* <span style="color:orange">PassengerId</span>: An integer identifier for each passenger. This is not incredibly relevant for analysis.

* <span style="color:blue">Survived</span>: A boolean detailing whether the passenger survived or not, 1 means they survived, 0 if they did not. This will be relevant for analysis

* <span style="color:blue">Pclass</span>: An Integer that indicates what class the passenger between 1st, 2nd and 3rd class is, 1st being the highest, 3rd being lowest. This will be relevant for analysis.

* <span style="color:red">Name</span>: A string that details the name of each passenger. This will not be relevant for analysis.

* <span style="color:blue">Sex</span>: A string that details the sex of the passenger between male and female. This will be relevant for analysis.

* <span style="color:blue">Age</span>: A double that details the age of each passenger. This will be relevant for analysis.

* <span style="color:blue">SibSp</span>: An integer detailing how many siblings and spouses each passenger had onboard. This will be relevant to the analysis.

* <span style="color:blue">Parch</span>: An integer that details the number of parents or children a passenger has onboard. This will be relevant to the analysis.

* <span style="color:red">Ticket</span>: A string that details the ID of the ticket used by the passenger. This will not be relevant for analysis.

* <span style="color:orange">Fare</span>: A double that details the money they paid to board the Titanic. This may be relevant to the analysis.

* <span style="color:red">Cabin</span>: A string that details what cabin they were assigned to. This is not relevant to the analysis.

* <span style="color:blue">Embarked</span>: A character detailing where the passenger boarded the ship from, S for Southampton, C for Cherbourg and Q for Queenstown. This will be relevant to the analysis.
